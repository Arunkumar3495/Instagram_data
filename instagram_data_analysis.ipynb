```python
# Importing the necessary python libries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud, STOPWORDS
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveRegressor
```

```python
df = pd.read_csv('/content/Instagram data.csv',encoding = 'latin1')
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.title('Distribution of impressions From Home')
sns.distplot(df['From Home'])
plt.show()
```

```python
def column_name():
  plt.figure(figsize=(10,5))
  plt.title('Distribution of impressions ')
  sns.distplot(df['column_name'])
  plt.show()
  return column_name
```

```python
plt.figure(figsize=(10,5))
plt.title('Distribution of impressions From Hashtags')
sns.distplot(df['From Hashtags'])
plt.show()
```

Hashtags are tool we use to categorize our post on instagram so that we can reach more people based on the kind of content. we are greeting looking at hashtags impression show that not all posts, can be reached using hashtags, but many new users can be reached from hashtags.

```python
plt.figure(figsize=(10,8))
plt.style.use('fivethirtyeight')
plt.title('Distribution of Impression From Explore')
sns.distplot(df['From Explore'])
plt.show()
```

The explore section of instagram is the recommendation system of instagram. Its recommended posts to the user based on their prefences and interests. Bylooking at the impression, I have received from the explore selection, I can say that instagram does not recoomened our posts much to the users. some post have received a good reach from the explore selection.

```python
home = df['From Home'].sum()
hashtags = df['From Hashtags'].sum()
explore = df['From Explore'].sum()
other = df['From Other'].sum()

labels = ['From Home','From Hashtags','From Explore','From Other']
values = [home,hashtags,explore,other]

fig = px.pie(df,values=values,names = labels,\
             title ='Impressions on Instagram Posts From Various Sources', hole=0.5)
fig.show()
```
